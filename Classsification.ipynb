```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder , OneHotEncoder
import pickle
```

```python
df = pd.read_csv(r"C:\Users\ammar\Downloads\Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace = True)
```

```python
df.head()
```

```python
label_encoder_gender = LabelEncoder()
df["Gender"] = label_encoder_gender.fit_transform(df["Gender"])
```

```python
df.head()
```

```python
one_hot_encode_geography = OneHotEncoder()
geography_encoded = one_hot_encode_geography.fit_transform(df[["Geography"]])
```

```python
geography_encoded.toarray()
```

```python
geo_encode = pd.DataFrame(geography_encoded.toarray(),columns=one_hot_encode_geography.get_feature_names_out(["Geography"]))
```

```python
geo_encode
```

```python
df = pd.concat([df,geo_encode],axis = 1)
```

```python
df.head()
```

```python
df.drop(columns = ['Geography'],inplace = True)
```

```python
df["Exited"].value_counts()
```

```python
df.head()
```

```python
X = df.drop(columns=["Exited"])
```

```python
y = df["Exited"]
```

```python
y.value_counts()
```

```python
X.head()
```

```python
standardscaler = StandardScaler()
X = standardscaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.2,random_state=12)
```

```python
from sklearn.utils.class_weight import compute_class_weight
classes = np.unique(y_train)
class_weight = compute_class_weight("balanced",classes=classes,y = y_train)
class_dict = dict(zip(classes,class_weight))
```

```python
with open("LabelEncoderGender.pkl","wb") as f:
    pickle.dump(label_encoder_gender,f)

with open("OneHotEncodergeography.pkl","wb") as f:
    pickle.dump(one_hot_encode_geography,f)

with open("StandardScaler.pkl","wb") as f:
    pickle.dump(standardscaler,f)
```

```python
X_train
```

```python
import tensorflow as tf
```

```python
X_train.shape[1]
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64,input_shape=(X_train.shape[1],)),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(32),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dense(16),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1,activation = "sigmoid")

])
```

```python
model.summary()
```

```python
loss = tf.keras.losses.binary_crossentropy
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=optimizer,loss = loss,metrics=["accuracy"])
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor="val_loss",patience=30,restore_best_weights=True)
```

```python
model.fit(X_train,y_train,epochs = 50,validation_data=(X_test,y_test),class_weight = class_dict,callbacks=[early_stop])
```
